```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Sample dataset: World Population
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Population': [7.383, 7.466, 7.549, 7.631, 7.713, 7.794, 7.874, 7.953, 8.030]
}

df = pd.DataFrame(data)

# Display the first few rows of the dataframe
print("First few rows of the dataset:")
print(df.head())

# Basic Statistics
print("\nBasic statistics of the dataset:")
print(df.describe())

# Plotting the data
plt.figure(figsize=(10, 12))

# Distribution of Population
plt.subplot(3, 1, 1)
plt.bar(df['Year'], df['Population'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.title('World Population Over Years')
plt.xticks(df['Year'])

# Line plot for Population growth over the years
plt.subplot(3, 1, 2)
plt.plot(df['Year'], df['Population'], marker='o', color='green')
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.title('Population Growth Over Years')
plt.xticks(df['Year'])

# Correlation analysis
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
print("\nCorrelation matrix:")
print(corr)

plt.subplot(3, 1, 3)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')

plt.tight_layout()
plt.show()
```
